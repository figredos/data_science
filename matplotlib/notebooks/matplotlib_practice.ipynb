```python
# Create a Python dictionary of 3 foods with the keys of the dictionary as the name and values the price
foods = {
    'pizza': 10,
    'fries': 5,
    'burger': 7,
}
```

```python
# Create a bar graph where the x-axis is the keys of the dictionary and y-axis as the values
fig, ax = plt.subplots()

ax.bar(foods.keys(), foods.values())

# Add a title, xlabel and ylabel to the pot
ax.set(title='Food Prices', xlabel='Food type', ylabel='Price')
```

```python
# Make the same plot as above, but with the bars horizontal
fig, ax = plt.subplots()

ax.barh(list(foods.keys()), list(foods.values()))

ax.set(title='Food Prices', ylabel='Food Types', xlabel='Prices')
```

```python
# Import pandas
import pandas as pd
```

```python
# Import the car_sales dataframe
car_sales = pd.read_csv('datasets/car-sales.csv')
```

```python
# Try to plot the 'Price' column using the plot() function
car_sales['Price'].plot()
```

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales['Price'] = car_sales['Price'].str.replace('[,.$]', '', regex=True)
car_sales['Price'] = car_sales['Price'].str[:-2].astype(int)
car_sales
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today
car_sales['Sale Date'] = pd.date_range('17/05/2024', periods=len(car_sales))

# View the car_sales DF
car_sales
```

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x='Sale Date', y='Total Sales')
```

```python
# Create the scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales.plot.scatter(x='Odometer (KM)', y='Price')
```

```python
# Create another df from the heart-disease dataset
heart_disease = pd.read_csv('datasets/heart-disease.csv')
```

```python
# View the first 10 rows of the heart_disease DF
heart_disease.head(10)
```

```python
# Create a histogram of the 'age' column with 50 bins
heart_disease['age'].plot.hist(bins=50)
```

```python
# Call plot.hist() on the heart_disease DF and toggle the subplots parameter to true
heart_disease.plot.hist(subplots=True)
```

```python
# Change the figsize of the plot above to (10, 30)
heart_disease.plot.hist(subplots=True, figsize=(10, 30))
```

```python
# Create a DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
scatter = ax.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
)

# Customize the plot
ax.set(
    title='Cholesterol vs Age analysis',
    xlabel='Age',
    ylabel='Cholesterol level',
)

ax.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
# Replot the same figure as above except change the 'cmap' parameter of scatter() to 'winter'
# Create a DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

# Customize the plot
ax.set(
    title='Cholesterol vs Age analysis',
    xlabel='Age',
    ylabel='Cholesterol level',
)

ax.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax.axhline(over_50['chol'].mean(), linestyle='--')
```
