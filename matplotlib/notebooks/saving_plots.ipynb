# Saving plots

Saving plots generated with pandas goes beyond aesthetics. While temporary plots disappear after script execution, saving them with Matplotlib's `plt.savefig()` or similar functions creates a permanent record. THis record serves multiple technical purposes. Firstly, it ensures reproducibility. Saved plots, along with the code, allow you or others to recreate the visualization using the same data, verifying the analysis.

Secondly, they empower effective dissemination of insights. Saved plots can be seamlessly integrated into reports, presentations or web applications enabling clear communications of complex data findings. Finally, version control of saved plots become crucial in collaborative data science projects. Integrating them with Git allows tracking changes over time, facilitating comparisons between iterations and collaborative refinement of visualizations.

In essence, saving plots transcends aesthetics, becoming a technical cornerstone for ensuring the longevity, replicability, and impactful communication of pandas-driven data analysis. It's also important to know that, the simplest way is to just 'save image as', this is possible because matplotlib actually generates the plot image as a png image.


```python
# Importing packages
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Creating Data
heart_disease = pd.read_csv('datasets/heart-disease.csv')
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
# Subplot of chol, age, thalcah
fig, (ax0, ax1) = plt.subplots(
    figsize=(10, 10),
    nrows=2,
    ncols=1,
    sharex=True,
)

# Adding data to ax0
scatter = ax0.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
    cmap='winter',
)

# Customize ax0
ax0.set(
    title='Heart Disease and Cholesterol levels',
    # xlabel='Age',
    ylabel='Cholesterol',
)

## Add a legend
ax0.legend(*scatter.legend_elements(), title='Target')

## Add horizontal line
ax0.axhline(
    over_50['chol'].mean(),
    linestyle='--',
    color='red',
)

## Adding data to ax1
scatter = ax1.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target'],
    cmap='winter',
)

# Customizing ax1
ax1.set(
    title='Heart disease and Max Heart Rate levels',
    xlabel='Age',
    ylabel='Max Heart Rate',
)

## Adding legend
ax1.legend(*scatter.legend_elements(), title='Target')

## Adding horizontal line
ax1.axhline(
    over_50['thalach'].mean(),
    linestyle='--',
    color='red',
)

## Adding a title to the figure
fig.suptitle('Heart Disease analysis', fontweight='bold')

# Limiting the axis
ax0.set_xlim([50, 80])
ax1.set_ylim([60, 200])
```

```python
# Saving image via code
fig.savefig('../images/heart_disease.png')
```
