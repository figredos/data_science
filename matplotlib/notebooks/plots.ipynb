# Plots

Plots are powerful tools for transforming raw data into easily digestible insights. They allow us to see patterns, trends, and relationships that might be difficult to grasp from tables of numbers.

Matplotlib empowers this process in several ways. First, it offers a broad range of plot types, from simple line charts to complex heat maps, allowing you to chose the best visualization to represent the given data. Second, Matplotlib's customization options enable you to tailor every aspect of your plot, from colors and markers to labels and titles.

Matplotlib offers many types of plots, such as line charts to track trends over time, scatter plots to reveal relationships between variables, bar charts for comparing categories, histograms to understand data distribution, and pie charts to visualize proportions of a whole. This diverse toolkit empowers you to choose the most effective visual representation to bring clarity and insights to your data.

```python
# Importing packages
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating data
x = np.linspace(0, 10, 100)
x[:10]
```

## Plot types


### Line Plot

Line plots are visual representation of data that use a number line and dots or markers. They're particularly useful for displaying discrete data, where values occur at specific points rather than continuously across a range.

Each dot on the line plot represents the frequency of a particular data value. For instance, a line plot might show the number of times a specific number appears in a data set. This makes them ideal for tasks like identifying the most frequent value (mode) or observing gaps or clusters in the data.

Line plots are generally simpler to interpret than other charts, and are often used for quick data exploration or to visualize small datasets.

The line plot may be the simplest one to create in Matplotlib, given that it's the default plot type generated by the `plt.plot()` function. This makes it a great starting point for visualizing trends or relationships between a single variable and time or another numerical variable.


```python
# Plotting the data and creating a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

### Scatter plot

Scatter plots, in contrast to line plots that tracks trends, are designed to reveal relationships between two numeric variables. Imagine a cloud of points where each dot represents a single data point. The horizontal position on the x-axis corresponds to the value of one variable, and the vertical position on the y-axis represents the value of the other variable.

By looking at the overall distribution of these dots, you can identify patters and potential correlations between the two variables. For example, a scatter plot might show weight on the y-axis and height of the x-axis. If the dots generally trend upwards, it suggests a positive correlation between height and weight.

Scatter plots are a powerful tool for exploring relationships and potential dependencies within your data, but it's important to remember that correlation doesn't necessarily imply causation.


```python
# Using the same data to create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

### Bar plot

Bar plots are workhorses for visualizing comparisons between categorical data. Unlike line plots, that deal with continuous data, bar charts shine when you have distinct categories, like types of products sold or customer age groups.

Each category is represented by a rectangular bar, with the bar's height or length proportional to the value it represents. This allows for easy visual comparisons between categories. For instance, a bar chart might show the number of customers in different age groups. The tallest bar would indicate the age group with the most customers.

Bar plots are great for quickly grasping the distribution of data across categories and identifying which category has the highest or lowest value. They can also be stacked or grouped to show additional layers of information within the same chart.

These plots can either be vertical, which they are by default when using `.bar`, or they can be horizontal, by using the `.barh` method. These two produce the same graphic, with the only difference between them being the disposition of the bars.


```python
# Make a bar plot from dictionary
nut_butter_prices = {
    'Almond Butter': 10,
    'Peanut Butter': 8,
    'Cashew Butter': 12,
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Matheus\'s Nut Butter Store', ylabel='Price ($)')
```

```python
# Creating horizontal bar plot from dictionary
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

### Histogram

Histograms are great at revealing the distribution of continuous data. Imagine having a collection of numerical values, like exam scores or house prices. While raw numbers can be overwhelming, a histogram transforms them into a clear visual story. It divides the entire range of values into smaller intervals, like bins.

Then, it counts how many data points fall within each bin. The result is a series of bars, where the height of each bar represents the frequency of data points within that particular value range. By looking at the histogram's shape, you can see if the data is clustered around a central value (normal distribution), skewed towards one end (positive or negative skew), or has multiple peaks (multimodal distribution).

Histograms are powerful tools for understanding the spread and overall shape of the data, making them invaluable for tasks like data cleaning, identifying outliers, and exploring potential relationships between variables.


```python
# Plotting normal distribution in a histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

### Pie plot

Pie charts, though widely used, offer a specific functionality in data visualization. They excel at portraying the composition of a whole, where each data slice represents a category and its proportional contribution to the total. Imagine a pie cut into widget, where the size of each wedge reflects the relative size of a category within the data set.

For instance, a pie chart might show the breakdown of spending categories in a budget, with each wedge representing the percentage spent on food, rent, transportation, and so on. Pie charts are effective for conveying proportions at a glance, but they can become cluttered and difficult to interpret when there are too many categories (more than 4-5 slices).

In such cases, alternative visualizations like bar charts or stacked bar charts might be more suitable.


```python
# Creating pie plot
fig, ax = plt.subplots()
ax.pie(nut_butter_prices.values(), labels=nut_butter_prices.keys())
```

### Box plot

Box plots offer a concise way to summarize the distribution of a dataset and identify potential outliers. Unlike most plot types, boxplots don't rely on individual data points for visualization. Instead, they focus on five key statistics, the minimum, maximum, median and first and third quartiles.

The box itself represents the interquartile range (IQR), encompassing the middle 50% of the data between the first and third quartiles. Lines extend rom the box (whiskers) to depict the range of data within 1.5 times the IQR from the quartiles. Any data points falling outside this whisker range are considered outliers and are plotted individually.

This approach allows you to quickly see how spread out the data is, whether it's symmetrical or skewed, and identify any extreme values that might warrant further investigation. Boxplots are particularly useful for comparing distributions across multiple datasets, helping you to spot differences in central tendency, variability, and the presence of outliers between groups.


```python
# Creating a box plot
fig, ax = plt.subplots()
ax.boxplot(x)
```

## SubPlots

Subplots are a powerful feature in Matplotlib that allow the creation of multiple plots within a single figure. Imagine a canvas divided into sections, where each section can hold a different visualization. This is particularly useful when you want to compare related data sets, visualize different aspects of the same data, or explore various relationships simultaneously.

Matplotlib provides functions like `plt.suplots()` to easily define a grid layout for the subplots, specifying the number of rows and columns. You can then treat each subplot like an individual plotting area, using Matplotlib's functionalities to create appropriate charts (line, plots, scatter plots, etc.) within each section.

This ability to combine multiple plots in a single figure helps creating more informative and cohesive visualizations, saving space and allowing viewers to easily make connections between the different panels.

There are two ways of creating subplots, you can either define the names for each plot individually when creating the subplots, or simply create as is and use index to reference each plot.


```python
# Option 1
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax0.plot(x, x / 2)

ax1.scatter(np.random.random(10), np.random.random(10))

ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax3.hist(np.random.randn(1000))
```

```python
# Option 2
fig, ax = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 5),
)

# Plot each different index
ax[0, 0].plot(x, x / 2)

ax[0, 1].scatter(np.random.random(10), np.random.random(10))

ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax[1, 1].hist(np.random.randn(1000))
```
