## Showing simple plots

Matplotlib offers two main approaches to plotting, pyplot and object-oriented. While both achieve the same goal, they differ in their approach and offer distinct advantages depending on the needs.

### Pyplot

Pyplot offers a simpler and more concise way to generate basic plots. Inspired by MATLAB, it uses a state-based approach. Functions like `plt.plot(x, y)` create a plot and modify the current figure on the board. Labels can be added with `plt.xlabel` and titles with `plt.title`, all building on the existing visualization. This approach is straightforward and perfect for quickly generating basic plots to explore data.

This simplicity however, comes with a cost. Since it relies on the current state, it can become cumbersome for complex plots. If you need to create two different subplots with different scales or customize line styles for each plot. Pyplot requires keeping track of the current state and making changes accordingly, which can lead to confusing code.

### Object-Oriented

Object-Oriented plotting offers a more granular and organized approach. Instead of a single state, you can explicitly create objects like `Figure` and `Axes`. With `plt.subplots()`, you can define the layout of plots (e.g., one figure with two subplots). They use methods on these objects to customize them. For instance, `fig.suptitle('My plot')` sets a title for the entire figure, and `ax1.plot(x1, y1, label='Data 1')` creates a line plot with a label on a specific subplot (`ax1`).

The object-oriented approach offers greater control and flexibility. It's particularly beneficial when creating complex plots with multiple elements or requiring extensive customization. It also promotes cleaner and more maintainable code, making it easier to understand and modify future visualizations.

### OBS.:

When creating plots, prefer using `plt.show()` to show them independently of them being pyplot or object oriented plots. This prevents the log of the object type in the output, showing instead the plot itself. An alternative to this, is always setting a semi-colon at the end of the plot creation `;`.


```python
# Creating empty plot
plt.plot()
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
# Creating lists of data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

```python
plt.plot(x, y)
plt.show()
```

```python
# 1st method
fig = plt.figure()
axes = fig.add_subplot()
plt.show()
```

```python
# 2nd method
fig = plt.figure()  # Creates a figure
axes = fig.add_axes((1, 1, 1, 1))
axes.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, axes = plt.subplots()
axes.plot(x, [59, 100, 200, 250])
plt.show()
```

# Anatomy of plot



As shown in the image above, a plot is composed of a series of smaller elements, these are some of them

### Data

This is the foundation of any plot. It's the numerical information to be represented visually.

### Axes

These are usually horizontal (x-axis) and vertical (y-axis) lines that establish the space where your data points will be plotted. The axes will have labels that indicate the units oor categories being measured. The scales on the axes define the range of values being visualized. They should be clear and consistent to allow viewers to accurately interpret the data.

### Ticks and Tick Labels

These represent the actual values from the data set. Depending on the plot type, they might be displayed as dots, bars, lines or other shapes.

### Markers and Colors

Markers like circles, squares, or triangles can be used to distinguish between different data sets plotted on the same axes. Colors can also be used effectively to differentiate data categories.

### Legend

If there are multiple data sets plotted on the same axes, a legend is crucial. It identifies the markers, colors, or line styles used for each data set, making the plot easier to interpret.

### Title

A clear and concise title positioned at the top of the plot briefly explains what the visualization is about.

### Figure

Figures represent the entire visualization, encompassing all the subplots, axes, data elements, and other visual components that come together.


## Matplotlib example workflow


```python
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize Plot
ax.set(
    title="Simple Plot",
    xlabel="x-axis",
    ylabel="y-axis",
)

# 5. Save and show
fig.savefig('../images/plot_1.png')
```
